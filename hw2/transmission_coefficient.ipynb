```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_e_v0 = lambda e_v0: (2*np.sqrt(e_v0)/(np.sqrt(e_v0) + np.sqrt(e_v0-1)))**2 * np.sqrt(e_v0-1) / np.sqrt(e_v0)
```

```python
x = np.linspace(1, 3)
plt.plot(x, t_e_v0(x), label='T')
plt.plot(x, 1-t_e_v0(x), label='R')
plt.xlabel('$E/V_0$')
plt.ylabel('Coefficient')
plt.legend()
plt.show()
```

```python
r_e_e0 = lambda e_e0: 1/(e_e0+1)
```

```python
x = np.linspace(0, 30)
plt.plot(x, 1-r_e_e0(x), label='T')
plt.plot(x, r_e_e0(x), label='R')
plt.xlabel('$E/E_0$')
plt.ylabel('Coefficient')
plt.legend()
plt.show()
```
