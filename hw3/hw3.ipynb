```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
```

```python
kb = 8.617E-5 # eV/K
mu = 1 # eV
f = lambda e, t: 1/(np.exp((e-mu)/(kb*t)) + 1)
```

```python
n = 100
ts = [0.01, 300, 1000, 2000, 5000, 9000, 20000]
es = np.linspace(0, 1, num=n)

result = np.zeros([n, len(ts)])
for i, temp in enumerate(ts):
    result[:, i] = f(es, temp)
```

```python
plt.plot(es, result)
plt.xlabel('E(eV)')
plt.ylabel('f(E,$\mu$,T)')
plt.show()
```

Problem 3 (a)

\begin{align}
    D(E) &= \frac{1}{A} \frac{dN(E)}{dE} \\
    &= \frac{1}{A} \frac{d}{dE} \bigg(\frac{L^2}{4\pi}\bigg) k^2 \\
    &= \frac{1}{A} \frac{d}{dE} \bigg(\frac{L^2}{4\pi}\bigg) \bigg( \frac{E}{\hbar v_{f}} \bigg)^2 \\
    &= \frac{1}{A} \frac{L^2}{2\pi \hbar^2 v_{f}^2} E
\end{align}

The $L^2$ term is the area of the sample. So $A$ and $L^2$ canceled out. Also the $E(k)$ can be either positive
or negative, but the density of states cannot be negative. So we multiply it by $2$, and take $E$ to absolute value.
Therefore, the density of states per spin of graphene is given by

\begin{align}
    D(E) &= \frac{1}{\pi \hbar^2 v_{f}^2} |E|
\end{align}

```python
nu_f = 1E8
e_3b = 100E-3 # eV
hbar = sc.physical_constants['reduced Planck constant in eV s'][0]
d_e = (np.pi*(hbar**2)*(nu_f**2))**(-1)*np.abs(e_3b)
print(f'The value of D(E) is {d_e:.2e} (eV cm^2)^-1.')
```
